# Project 3 - SQL Data Analysis

Dataset loaded from Excel into SQLite and analyzed using SQL queries.

```python
import pandas as pd
import sqlite3

df = pd.read_excel('Dataset for Data Analytics.xlsx')
conn = sqlite3.connect(':memory:')
df.to_sql('orders', conn, index=False, if_exists='replace')
print('Rows:', len(df))
df.head()
```

## 1. SELECT Query

```python
pd.read_sql('SELECT OrderID, Product, TotalPrice FROM orders LIMIT 10', conn)
```

## 2. WHERE Clause

```python
pd.read_sql("SELECT * FROM orders WHERE TotalPrice > 2000 LIMIT 10", conn)
```

## 3. ORDER BY

```python
pd.read_sql('SELECT OrderID, Product, TotalPrice FROM orders ORDER BY TotalPrice DESC LIMIT 10', conn)
```

## 4. GROUP BY with COUNT

```python
pd.read_sql('SELECT PaymentMethod, COUNT(*) AS TotalOrders FROM orders GROUP BY PaymentMethod ORDER BY TotalOrders DESC', conn)
```

## 5. SUM Aggregation

```python
pd.read_sql('SELECT OrderStatus, SUM(TotalPrice) AS Revenue FROM orders GROUP BY OrderStatus', conn)
```

## 6. AVG Aggregation

```python
pd.read_sql('SELECT Product, AVG(TotalPrice) AS AvgPrice FROM orders GROUP BY Product ORDER BY AvgPrice DESC', conn)
```

## 7. Combined Analysis (WHERE + GROUP BY + ORDER BY)

```python
pd.read_sql("""
SELECT ReferralSource,
COUNT(*) AS Orders,
ROUND(AVG(TotalPrice),2) AS AvgOrderValue
FROM orders
WHERE TotalPrice > 500
GROUP BY ReferralSource
ORDER BY AvgOrderValue DESC
""", conn)
```
