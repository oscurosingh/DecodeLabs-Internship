## 4. 🗄️ Create SQLite Database

We load the DataFrame into an **in-memory SQLite database**. This replicates a production relational database environment and allows us to execute standard SQL queries exactly as we would in MySQL, PostgreSQL, or SQL Server.


````python
# Create in-memory SQLite database
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

# Load the DataFrame into a SQL table named 'orders'
df.to_sql('orders', conn, if_exists='replace', index=False)

cursor.execute('SELECT COUNT(*) FROM orders')
row_count = cursor.fetchone()[0]

print('=' * 50)
print('  SQLite Database Initialised')
print('=' * 50)
print(f'  Table Name  : orders')
print(f'  Total Rows  : {row_count:,}')
print(f'  Status      : ✅ Ready for SQL queries')
print('=' * 50)
````

### Helper Function — Clean Query Execution

````python
def run_query(sql, title=None):
    """Execute SQL and return results as a formatted DataFrame."""
    result = pd.read_sql_query(sql, conn)
    if title:
        print(f'\n📊  {title}')
        print('-' * 60)
    return result

print('✅ Query helper function ready.')
````

## 5. 🔍 Basic SELECT Queries

These foundational queries retrieve raw data from the database without any filtering or aggregation. They form the backbone of every analytical workflow.


### Query 1 — View All Columns (First 10 Records)

````python
sql = """
SELECT *
FROM   orders
LIMIT  10;
"""

run_query(sql, 'All Columns — First 10 Records')
````

> **📌 Business Insight:** A full-column preview confirms the data loaded correctly across all 14 attributes and allows a quick sanity check on data types, value ranges, and potential anomalies before deeper analysis begins.


### Query 2 — Select Key Business Columns Only

````python
sql = """
SELECT OrderID,
       Date,
       Product,
       Quantity,
       UnitPrice,
       TotalPrice,
       OrderStatus
FROM   orders
LIMIT  10;
"""

run_query(sql, 'Core Business Fields')
````

> **📌 Business Insight:** In production analytics, selecting only needed columns reduces I/O and processing time — a critical habit when querying tables with millions of rows. This query mirrors what a BI dashboard data layer would fetch.


### Query 3 — Distinct Product Categories in Catalogue

````python
sql = """
SELECT DISTINCT Product
FROM   orders
ORDER  BY Product;
"""

run_query(sql, 'Unique Product Categories')
````

> **📌 Business Insight:** 7 product categories span electronics (Laptop, Phone, Tablet, Monitor, Printer) and furniture (Chair, Desk), indicating a diversified product strategy. Understanding catalogue breadth is step one of any category management analysis.


### Query 4 — Distinct Payment Methods Accepted

````python
sql = """
SELECT DISTINCT PaymentMethod
FROM   orders
ORDER  BY PaymentMethod;
"""

run_query(sql, 'Available Payment Methods')
````

### Query 5 — Distinct Order Status Pipeline

````python
sql = """
SELECT DISTINCT OrderStatus
FROM   orders
ORDER  BY OrderStatus;
"""

run_query(sql, 'Order Status Pipeline')
````

> **📌 Business Insight:** Five status stages capture the complete order lifecycle. `Cancelled` and `Returned` orders represent direct revenue leakage and should be monitored closely in operational reporting dashboards.

## 6. 🔎 WHERE Clause — Filtering & Segmentation

The `WHERE` clause is the primary filtering mechanism in SQL. It evaluates every individual row **before** any grouping or aggregation occurs (execution order step 2), making early, precise filtering critical for both query performance and analytical accuracy.


### Query 6 — High-Value Orders (TotalPrice > ₹2,000)

````python
sql = """
SELECT OrderID,
       Product,
       Quantity,
       TotalPrice,
       PaymentMethod,
       OrderStatus
FROM   orders
WHERE  TotalPrice > 2000
ORDER  BY TotalPrice DESC
LIMIT  15;
"""

run_query(sql, 'High-Value Transactions (TotalPrice > 2000)')
````

> **📌 Business Insight:** High-value orders represent premium customers who are strong candidates for a loyalty or VIP programme. These transactions deserve priority in retention strategies, personalised outreach, and white-glove fulfilment.


### Query 7 — All Laptop Orders

````python
sql = """
SELECT OrderID,
       Date,
       CustomerID,
       Quantity,
       UnitPrice,
       TotalPrice,
       OrderStatus
FROM   orders
WHERE  Product = 'Laptop'
LIMIT  10;
"""

run_query(sql, 'All Laptop Orders')
````

> **📌 Business Insight:** Isolating a single product's orders enables the procurement team to assess demand volume independently and align inventory restocking cycles. Laptops typically carry the highest unit price, making demand forecasting critical.


### Query 8 — Delivered Orders (Confirmed Revenue)

````python
sql = """
SELECT OrderID,
       Product,
       TotalPrice,
       PaymentMethod,
       ReferralSource
FROM   orders
WHERE  OrderStatus = 'Delivered'
LIMIT  10;
"""

run_query(sql, 'Successfully Delivered Orders')
````

> **📌 Business Insight:** Delivered orders represent confirmed, realised revenue. Analysing this subset reveals which products and acquisition channels consistently drive successful fulfilment — the gold standard for financial reporting.


### Query 9 — Multi-Condition Filter: Delivered Laptops Only

````python
sql = """
SELECT OrderID,
       Product,
       Quantity,
       TotalPrice,
       OrderStatus,
       PaymentMethod
FROM   orders
WHERE  Product      = 'Laptop'
  AND  OrderStatus  = 'Delivered'
ORDER  BY TotalPrice DESC;
"""

run_query(sql, 'Delivered Laptop Orders')
````

> **📌 Business Insight:** Combining multiple WHERE conditions creates precise analytical segments. Filtering to delivered laptop orders specifically yields the laptop category's *realised* revenue — essential for product-level P&L analysis.


### Query 10 — Google-Referred Customer Orders

````python
sql = """
SELECT OrderID,
       Product,
       TotalPrice,
       ReferralSource,
       PaymentMethod,
       OrderStatus
FROM   orders
WHERE  ReferralSource = 'Google'
LIMIT  12;
"""

run_query(sql, 'Google-Referred Customer Orders')
````

> **📌 Business Insight:** Google referral data enables precise ROI measurement for search engine marketing (SEM) spend by mapping actual order values to that acquisition channel — the foundation of any marketing attribution model.


### Query 11 — Cancelled Orders (Revenue Leakage Analysis)

````python
sql = """
SELECT OrderID,
       Product,
       TotalPrice,
       PaymentMethod,
       ReferralSource
FROM   orders
WHERE  OrderStatus = 'Cancelled'
ORDER  BY TotalPrice DESC
LIMIT  10;
"""

run_query(sql, 'Cancelled Orders — Revenue at Risk')
````

> **📌 Business Insight:** High-value cancellations represent direct, measurable revenue loss. Root cause analysis on cancellation reasons — payment failure, stock-outs, customer indecision — is essential for revenue recovery and checkout optimisation.

## 7. 📊 ORDER BY — Sorting & Rankings

`ORDER BY` operates at execution step 6 — the **final** step after all filtering and aggregation — making it the presentation layer of any SQL query. It can sort by column aliases defined in `SELECT`, which is why it works even when `WHERE` cannot reference those same aliases.


### Query 12 — Top 10 Highest-Revenue Orders

````python
sql = """
SELECT OrderID,
       Product,
       Quantity,
       UnitPrice,
       TotalPrice,
       PaymentMethod
FROM   orders
ORDER  BY TotalPrice DESC
LIMIT  10;
"""

run_query(sql, 'Top 10 Highest-Value Orders')
````

> **📌 Business Insight:** The top 10 orders by revenue often reveal which product categories dominate the premium customer segment. A recurring product in this list indicates a high-margin SKU worthy of dedicated inventory investment and VIP customer treatment.


### Query 13 — Bottom 10 Lowest-Value Orders

````python
sql = """
SELECT OrderID,
       Product,
       Quantity,
       UnitPrice,
       TotalPrice
FROM   orders
ORDER  BY TotalPrice ASC
LIMIT  10;
"""

run_query(sql, '10 Lowest-Value Orders')
````

> **📌 Business Insight:** Extremely low-value orders may have negative unit economics after accounting for shipping, packaging, and handling costs. This data is critical for establishing minimum order value policies or free-shipping thresholds.


### Query 14 — Most Recent Orders (Date Sort)

````python
sql = """
SELECT OrderID,
       Date,
       Product,
       TotalPrice,
       OrderStatus
FROM   orders
ORDER  BY Date DESC
LIMIT  10;
"""

run_query(sql, '10 Most Recent Orders')
````

> **📌 Business Insight:** Sorting by date descending provides the operations team with a real-time order feed, allowing rapid identification of recently placed orders that may require immediate processing or fulfilment attention.


### Query 15 — Products Alphabetically Sorted

````python
sql = """
SELECT DISTINCT Product
FROM   orders
ORDER  BY Product ASC;
"""

run_query(sql, 'Product Catalogue — Alphabetical Order')
````

## 8. 📦 GROUP BY & Aggregations — COUNT / SUM / AVG

`GROUP BY` collapses individual transaction rows into categorical **buckets**. Combined with aggregate functions, it transforms 1,200 raw records into concise executive-level KPIs. This is the mechanism behind every chart, table, and metric in a business intelligence dashboard.


### Query 16 — COUNT: Total Transactions per Product

````python
sql = """
SELECT Product,
       COUNT(*) AS Total_Orders
FROM   orders
GROUP  BY Product
ORDER  BY Total_Orders DESC;
"""

run_query(sql, 'Order Volume by Product')
````

> **📌 Business Insight:** `COUNT(*)` measures demand frequency, not revenue magnitude. A high-volume but low-ticket product (e.g., Desk) signals different strategic needs than a low-volume, high-ticket product (e.g., Laptop). Both dimensions are required for complete portfolio analysis.


### Query 17 — COUNT: Transactions by Payment Method

````python
sql = """
SELECT PaymentMethod,
       COUNT(*) AS Transaction_Count
FROM   orders
GROUP  BY PaymentMethod
ORDER  BY Transaction_Count DESC;
"""

run_query(sql, 'Transaction Volume by Payment Method')
````

> **📌 Business Insight:** The dominant payment method determines which payment gateway deserves the highest reliability investment. A checkout breakdown on the top payment method has the highest conversion impact of any technical failure on the platform.


### Query 18 — COUNT + Percentage: Order Status Distribution

````python
sql = """
SELECT OrderStatus,
       COUNT(*) AS Order_Count,
       ROUND(COUNT(*) * 100.0 / (SELECT COUNT(*) FROM orders), 2) AS Pct_Share
FROM   orders
GROUP  BY OrderStatus
ORDER  BY Order_Count DESC;
"""

run_query(sql, 'Order Status Distribution with Percentage Share')
````

> **📌 Business Insight:** The percentage share of each order status is a key operational health metric. A combined cancellation + return rate exceeding 20% signals product quality issues, misleading listings, or payment gateway problems requiring immediate investigation.


### Query 19 — SUM: Total Revenue by Product

````python
sql = """
SELECT Product,
       SUM(TotalPrice)                         AS Total_Revenue,
       COUNT(*)                                AS Order_Count,
       ROUND(SUM(TotalPrice) / COUNT(*), 2)    AS Avg_Order_Value
FROM   orders
GROUP  BY Product
ORDER  BY Total_Revenue DESC;
"""

run_query(sql, 'Revenue Summary by Product')
````

> **📌 Business Insight:** `SUM(TotalPrice)` reveals actual revenue contribution per category. Presenting Total Revenue alongside Order Count and AOV in the same query gives a complete product performance view — volume, value, and efficiency in a single result set.


### Query 20 — SUM: Total Revenue by Payment Method

````python
sql = """
SELECT PaymentMethod,
       COUNT(*)                  AS Transactions,
       ROUND(SUM(TotalPrice), 2) AS Total_Revenue,
       ROUND(AVG(TotalPrice), 2) AS Avg_Order_Value
FROM   orders
GROUP  BY PaymentMethod
ORDER  BY Total_Revenue DESC;
"""

run_query(sql, 'Revenue by Payment Method')
````

> **📌 Business Insight:** Comparing revenue per payment method vs. transaction count reveals average ticket size differences across payment types. This metric helps the product team prioritise which payment flows to optimise for revenue maximisation vs. volume maximisation.


### Query 21 — AVG: Average Order Value with Min/Max by Product

````python
sql = """
SELECT Product,
       ROUND(AVG(TotalPrice), 2)  AS Avg_Order_Value,
       ROUND(MIN(TotalPrice), 2)  AS Min_Order_Value,
       ROUND(MAX(TotalPrice), 2)  AS Max_Order_Value
FROM   orders
GROUP  BY Product
ORDER  BY Avg_Order_Value DESC;
"""

run_query(sql, 'Average Order Value Statistics by Product')
````

> **📌 Business Insight:** Average Order Value (AOV) is a primary pricing benchmark. Products with high AOV but moderate volume are ideal upselling targets because each incremental conversion carries disproportionately high revenue impact compared to lower-AOV categories.


### Query 22 — AVG + SUM: Marketing Channel Performance

````python
sql = """
SELECT ReferralSource,
       COUNT(*)                    AS Orders,
       ROUND(AVG(TotalPrice), 2)   AS Avg_Order_Value,
       ROUND(SUM(TotalPrice), 2)   AS Total_Revenue
FROM   orders
GROUP  BY ReferralSource
ORDER  BY Total_Revenue DESC;
"""

run_query(sql, 'Performance by Marketing / Referral Channel')
````

> **📌 Business Insight:** This is a Channel Attribution Report. Marketing teams use AOV per channel to compute Customer Acquisition Cost (CAC) efficiency — a channel with higher AOV justifies higher ad spend per click because each conversion returns more revenue.


### Query 23 — AVG & SUM: Coupon Code Performance

````python
sql = """
SELECT CouponCode,
       COUNT(*)                    AS Times_Used,
       ROUND(SUM(TotalPrice), 2)   AS Total_Revenue,
       ROUND(AVG(TotalPrice), 2)   AS Avg_Order_Value
FROM   orders
WHERE  CouponCode IS NOT NULL
GROUP  BY CouponCode
ORDER  BY Total_Revenue DESC;
"""

run_query(sql, 'Coupon Code Performance Analysis')
````

> **📌 Business Insight:** Coupon performance analysis validates promotion ROI. If coupon users have a higher AOV than the platform average, it confirms that discounts act as basket-building tools — increasing cart size — rather than simply reducing margins on existing intent.

## 9. 🚀 Advanced SQL Analysis — Multi-Clause Business Intelligence

These queries combine `WHERE`, `GROUP BY`, `ORDER BY`, `HAVING`, aggregate functions, and conditional logic (`CASE WHEN`) to generate executive-level insights that mirror real-world BI dashboard queries and analyst reports.


### Query 24 — Confirmed Revenue by Product (Delivered Orders Only)

````python
sql = """
SELECT Product,
       COUNT(*)                    AS Delivered_Orders,
       ROUND(SUM(TotalPrice), 2)   AS Confirmed_Revenue,
       ROUND(AVG(TotalPrice), 2)   AS Avg_Order_Value
FROM   orders
WHERE  OrderStatus = 'Delivered'
GROUP  BY Product
ORDER  BY Confirmed_Revenue DESC;
"""

run_query(sql, 'Confirmed Revenue (Delivered Orders Only) by Product')
````

> **📌 Business Insight:** Filtering to `Delivered` status before aggregating yields *confirmed net revenue* — the standard for financial reporting since Cancelled and Returned orders do not generate realised revenue. This is the query that feeds P&L statements and revenue recognition reports.


### Query 25 — Best Marketing Channel for High-Value Customers

````python
sql = """
SELECT ReferralSource,
       COUNT(*)                    AS High_Value_Orders,
       ROUND(SUM(TotalPrice), 2)   AS Revenue_From_Source,
       ROUND(AVG(TotalPrice), 2)   AS Avg_Order_Value
FROM   orders
WHERE  TotalPrice > 1000
GROUP  BY ReferralSource
ORDER  BY Revenue_From_Source DESC;
"""

run_query(sql, 'Best Marketing Channel for High-Value Orders (> ₹1,000)')
````

> **📌 Business Insight:** This directly answers the CMO's strategic question: *'Which acquisition channel brings our highest-value customers?'* — a decision that determines next quarter's media budget allocation across all channels.


### Query 26 — Payment Preferences by Product Category

````python
sql = """
SELECT Product,
       PaymentMethod,
       COUNT(*) AS Order_Count
FROM   orders
GROUP  BY Product, PaymentMethod
ORDER  BY Product, Order_Count DESC;
"""

run_query(sql, 'Payment Method Preference per Product')
````

> **📌 Business Insight:** Understanding payment preferences per product category enables the checkout engineering team to surface the statistically preferred payment option first for each product type — a proven conversion rate optimisation tactic that reduces payment friction.


### Query 27 — Cancellation Rate by Product (CASE WHEN)

````python
sql = """
SELECT Product,
       COUNT(*) AS Total_Orders,
       SUM(CASE WHEN OrderStatus = 'Cancelled' THEN 1 ELSE 0 END) AS Cancelled,
       SUM(CASE WHEN OrderStatus = 'Returned'  THEN 1 ELSE 0 END) AS Returned,
       ROUND(
           SUM(CASE WHEN OrderStatus IN ('Cancelled','Returned') THEN 1 ELSE 0 END)
           * 100.0 / COUNT(*), 2
       ) AS Attrition_Rate_Pct
FROM   orders
GROUP  BY Product
ORDER  BY Attrition_Rate_Pct DESC;
"""

run_query(sql, 'Revenue Attrition Rate (Cancelled + Returned) by Product')
````

> **📌 Business Insight:** `CASE WHEN` enables conditional aggregation within a single GROUP BY query. Products with attrition rates significantly above the category average may have misleading listings, pricing inconsistencies, or quality issues requiring catalogue review and customer feedback analysis.


### Query 28 — Monthly Revenue Trend (Time-Series Analysis)

````python
sql = """
SELECT STRFTIME('%Y-%m', Date) AS Year_Month,
       COUNT(*)                        AS Orders,
       ROUND(SUM(TotalPrice), 2)       AS Monthly_Revenue,
       ROUND(AVG(TotalPrice), 2)       AS Avg_Order_Value
FROM   orders
GROUP  BY Year_Month
ORDER  BY Year_Month;
"""

run_query(sql, 'Monthly Revenue Trend (2023–2024)')
````

> **📌 Business Insight:** Month-over-month revenue trends reveal seasonality patterns, peak sales periods, and revenue dips. This time-series data is the foundation for demand forecasting, inventory procurement planning, and marketing calendar design.


### Query 29 — Coupon + Delivered Revenue by Product (Combined Filters)

````python
sql = """
SELECT Product,
       CouponCode,
       COUNT(*)                    AS Orders_With_Coupon,
       ROUND(SUM(TotalPrice), 2)   AS Confirmed_Revenue,
       ROUND(AVG(TotalPrice), 2)   AS Avg_Order_Value
FROM   orders
WHERE  CouponCode   IS NOT NULL
  AND  OrderStatus  = 'Delivered'
GROUP  BY Product, CouponCode
ORDER  BY Confirmed_Revenue DESC
LIMIT  10;
"""

run_query(sql, 'Confirmed Revenue by Product + Coupon Combination')
````

> **📌 Business Insight:** This identifies winning product-coupon pairings by *confirmed* (delivered) revenue. The promotions team uses this data to double down on combinations that drive high-value completed purchases, rather than coupons that attract low-quality orders with high cancellation rates.


### Query 30 — Repeat Customers: Loyalty Signal Detection (HAVING)

````python
sql = """
SELECT CustomerID,
       COUNT(*)                    AS Total_Orders,
       ROUND(SUM(TotalPrice), 2)   AS Total_Lifetime_Spent,
       ROUND(AVG(TotalPrice), 2)   AS Avg_Order_Value
FROM   orders
GROUP  BY CustomerID
HAVING COUNT(*) > 1
ORDER  BY Total_Lifetime_Spent DESC
LIMIT  15;
"""

run_query(sql, 'Repeat Customers — Top 15 by Lifetime Value (HAVING clause)')
````

> **📌 Business Insight:** `HAVING` filters on aggregated results — impossible with `WHERE` alone because `WHERE` runs before grouping. Repeat customers have substantially lower CAC than new acquisitions and are prime candidates for loyalty programmes, personalised offers, and subscription product upgrades.
